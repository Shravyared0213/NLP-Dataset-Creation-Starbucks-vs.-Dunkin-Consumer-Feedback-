```python
from google_play_scraper import reviews, Sort
import pandas as pd

# Scraping 10000 reviews of Starbucks App from googleplaystore
result, _ = reviews(
    'com.starbucks.mobilecard',  # ← Starbucks App ID
    lang='en',                   # Language
    country='us',                # Country
    sort=Sort.NEWEST,            # Can also use Sort.MOST_RELEVANT
    count=10000                    # Number of reviews to scrape
)

# Convert to DataFrame
df = pd.DataFrame(result)

# Keep only needed columns
df = df[["userName", "score", "content", "at"]]
df.rename(columns={
    "userName": "username",
    "score": "rating",
    "content": "review",
    "at": "date"
}, inplace=True)



# Save the data into csv file
df.to_csv("starbucks_googleplay_reviews.csv", index=False)
df.head()
```

```python
df.head()
# Scraping 10000 reviews of Dunkin App from googleplaystore
result, _ = reviews(
    'com.dunkinbrands.otgo',  # ← Starbucks App ID
    lang='en',                   # Language
    country='us',                # Country
    sort=Sort.NEWEST,            # Can also use Sort.MOST_RELEVANT
    count=10000                    # Number of reviews to scrape
)

# Convert to DataFrame
df1 = pd.DataFrame(result)


# Keep only needed columns
df1 = df1[["userName", "score", "content", "at"]]
df1.rename(columns={
    "userName": "username",
    "score": "rating",
    "content": "review",
    "at": "date"
}, inplace=True)



# Save the data into csv file
df1.to_csv("Dunkin_googleplay_reviews.csv", index=False)
df1.head()
```

```python
import nltk
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
from wordcloud import WordCloud
from nltk.tokenize import word_tokenize

# Download missing resource
nltk.download('punkt_tab')

def clean_text(text):
  # Check if the input is a string, otherwise return an empty list
  if not isinstance(text, str):
    return []
  tokens = word_tokenize(text)
  # Convert to lowercase
  tokens = [word.lower() for word in tokens]
  # Remove punctuation and non-alphabetic characters
  tokens = [word for word in tokens if word.isalpha()]
  # Remove stop words
  stop_words = set(stopwords.words('english'))
  tokens = [word for word in tokens if word not in stop_words]
  # Lemmatize words
  lemmatizer = WordNetLemmatizer()
  tokens = [lemmatizer.lemmatize(word) for word in tokens]
  return tokens

# Example usage:
df3['cleaned_review'] = df3['review'].apply(clean_text)
display(df3.head())
```

```python
from wordcloud import WordCloud
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt
import os

# Join reviews
starbucks_text = " ".join([" ".join(review) for review in df3[df3['App']=='Starbucks']['cleaned_review']])
dunkin_text = " ".join([" ".join(review) for review in df3[df3['App']=='Dunkin']['cleaned_review']])

# Check if mask files exist
starbucks_mask_path = "starbucks.png"
dunkin_mask_path = "donut.png"

# Load, convert & binarize masks
mask_star = np.array(Image.open(starbucks_mask_path).convert("L"))
mask_star = np.where(mask_star < 128, 0, 255)

mask_dunk = np.array(Image.open(dunkin_mask_path).convert("L"))
mask_dunk = np.where(mask_dunk < 128, 0, 255)

    # Generate word clouds
wc_star = WordCloud(background_color='white',
                        mask=mask_star,
                        contour_width=2,
                        contour_color='green',
                        width=mask_star.shape[1],
                        height=mask_star.shape[0]).generate(starbucks_text)

wc_dunk = WordCloud(background_color='white',
                        mask=mask_dunk,
                        contour_width=2,
                        contour_color='orange',
                        width=mask_dunk.shape[1],
                        height=mask_dunk.shape[0]).generate(dunkin_text)

    # Plot
plt.figure(figsize=(14,6))
plt.subplot(1,2,1); plt.imshow(wc_star, interpolation='bilinear'); plt.axis('off'); plt.title("Starbucks")
plt.subplot(1,2,2); plt.imshow(wc_dunk, interpolation='bilinear'); plt.axis('off'); plt.title("Dunkin")
plt.show()
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix, classification_report,accuracy_score
import matplotlib.pyplot as plt
import seaborn as sns

# Drop rows with missing values in 'review' column
df_cleaned = df3.dropna(subset=['review'])

#vectorizing the reviews
cv=TfidfVectorizer(max_features=2500)
x=cv.fit_transform(df_cleaned['review']).toarray()
y=df_cleaned['sentiment']

#training the model
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=0)
model=LogisticRegression()
model.fit(x_train,y_train)
y_pred=model.predict(x_test)
#accuracy score
print(f"accuracy score: {accuracy_score(y_test,y_pred)}")
print(f"classification_report :\n{classification_report(y_test,y_pred)}")
print(f"confusion matrix:{confusion_matrix(y_test,y_pred)}")
#plotting the Confusion Matrix
cm= confusion_matrix(y_test,y_pred)
sns.heatmap(cm,annot=True,fmt="d",cmap='Greens')
plt.xlabel('predicted')
plt.ylabel('Actual')
plt.title('Actual VS prediction ')
plt.show()
```

```python
#adding predictions back to dataset
df_cleaned = df3.dropna(subset=['review'])
df_cleaned['predicted sentiment']=model.predict(cv.transform(df_cleaned['review']).toarray())
# Starbucks sentiment breakdown
starbucks_results = df_cleaned[df_cleaned['App'] == 'Starbucks']
print(f"starbucks prediction:{starbucks_results['predicted sentiment'].value_counts()}")
# Dunkin sentiment breakdown
dunkin_results = df_cleaned[df_cleaned['App'] == 'Dunkin']
print(f"dunkin prediction : {dunkin_results['predicted sentiment'].value_counts()}")
```

```python
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.utils import pad_sequences
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM,Dropout,Dense,Embedding
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Drop rows with missing values in 'review' column
df_cleaned = df3.dropna(subset=['review'])

y = df_cleaned['sentiment'].values
tokenize=Tokenizer(num_words=5000) # Set num_words to match embedding layer input_dim
tokenize.fit_on_texts(df_cleaned['review'])
x_seq=tokenize.texts_to_sequences(df_cleaned['review'])
x_padded=pad_sequences(x_seq,maxlen=200)

X_train, X_test, y_train, y_test = train_test_split(x_padded, y, test_size=0.2, random_state=42)


model_lstm = Sequential()
model_lstm.add(Embedding(input_dim=5000, output_dim=128))
model_lstm.add(LSTM(128, return_sequences=False))
model_lstm.add(Dropout(0.5))
model_lstm.add(Dense(1, activation='sigmoid'))

model_lstm.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
model_lstm.summary()
history = model_lstm.fit(X_train, y_train, epochs=5, batch_size=64, validation_data=(X_test, y_test))
loss, accuracy = model_lstm.evaluate(X_test, y_test)
print(f"LSTM Accuracy: {accuracy:.4f}")

# Predict
y_pred_lstm = (model_lstm.predict(X_test) > 0.5).astype("int32")

# Classification Report
print("LSTM Classification Report:\n", classification_report(y_test, y_pred_lstm))

print(f"confusion matrix:{confusion_matrix(y_test,y_pred_lstm)}")

# Confusion Matrix
cm_lstm = confusion_matrix(y_test, y_pred_lstm)
sns.heatmap(cm_lstm, annot=True, fmt='d', cmap='Blues')
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.title("LSTM Confusion Matrix")
plt.show()
```
